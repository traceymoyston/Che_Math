# Problem 1 Flow and Bifurcation

1. You are studying energy potential diagrams and come across the following equation:
$\frac{dx}{dt} = rx+x^3-x^5$. Let **r = -2** and use the code cell below to plot the flow figure. Be sure to label axes. Some quick hints:
- You will likely have to zoom into the figure to see the important points. Consider using ylim and xlim.
- It may help to draw an abline at y = 0. Look up the command "axhline".

```python
#plotting the flow diagrams for a differential equation with different parameters for comparison 


#imports the necessary packages to run the code 
import matplotlib.pyplot as plt 
import numpy as np 


def flowdx(x,t): #creates a function that finds the value of the differential equation for different x values 
    r=-2
    dxdt=r*x+x**3-x**5;
    return dxdt #returns the value of the equation at x 

x_vals=np.linspace(-2,2,50) #creates a vector of x values 


dx_dt=[]; #creates an empty vector to add in the differential equation values 

for i in x_vals: #loops through the x value list and evaluates the equation at this value 
    dx_dt.append(flowdx(i,x_vals))
    
# print(x_vals)
# print(dx_dt)

#plots the values and adds labels to the plot 
plt.plot(x_vals,dx_dt)
plt.axhline(color='r') #adds a red line at y=0 
plt.xlabel('x')
plt.ylabel('dx/dt')
plt.title('Flow diagram for the equation at r=-2' )
# plt.xlim([-0.1,0.1]) #zooms in to the plot to find where y=0 intersects with the plot 
# plt.ylim([-0.1,0.1])
    
```

In the cell below answer : How many steadystates are there and what are their *approximate* values (i.e. zoom into the figure and read them from the diagram)? Describe the stability of each steady state.


There is $1$ steadystate and it is at $x=0$. the steady state is unstable.

To ensure we understand what the flow diagram means in terms of the dynamic changes in the value of x over time, please use an ode solver to simulate the ODE with $x_0 = 0.8$. Plot the value of x versus time in the space below. For the simulation time and time step size, choose appropriate numbers such that the plot is smooth and the figure clearly shows the approach to steadystate.

```python
from scipy.integrate import odeint

timesteps = np.linspace(0,10,100) #list of time values 
x0 = 0.8  # the initial condition
x = odeint(flowdx, x0, timesteps); #solves the differential equation for the given initial condition 
plt.plot(timesteps,x)
plt.xlabel('t')
plt.ylabel('x')
plt.title('Plot of x vs t for x0=0.8 and r=-2' )
```

# Problem 2 Linearization

You're given the nonlinear, first-order, time-invariant ODE $\frac{dy}{dt} = sin(y)$.

1. In the space below, plot the flow diagram for the ODE. 

```python
#linearization of nonlinear ODE 

import math 

def flowdy(y,t): #creates a function to return the value of the ode for different values of y 
    dydt= math.sin(y);
    return dydt #returns the ode value 

y_vals= np.linspace(0,10,50) #list of y values 

dy_dt=[];
for i in y_vals:
    dy_dt.append(flowdy(i,0))

plt.plot(y_vals,dy_dt)
plt.axhline(color='r')
plt.xlabel('y')
plt.ylabel('dy/dt')
plt.title('Flow diagram for the equation' )
```

2. In the space below, type out the equation for the linearized form of our ODE when it is linearized about the lowest, positive-valued, stable steadystate.

$dy/dt=-y+\pi$

lowest positive stable steadystate is at $\pi$ where the derivative of that section would be negative 

3. We want to compare the responses of our linearized model to our original nonlinear model. On a single figure, plot the time response of both models when starting at an initial conditions of $y_0 = 3$. Clearly indicate which lines represent which ODEs.

```python
def lineardy(y,t): # solves the equation at different values of y 
    dydt=-y+math.pi
    return dydt 

timesteps=np.linspace(0,6,20)
y0=3; #initial condition 

y1= odeint(flowdy, y0, timesteps) #solves the differential equations for the initial condition above 
y2= odeint(lineardy, y0, timesteps)

#plots the two solutions on the same plot 
plt.plot(timesteps,y1)
plt.plot(timesteps,y2)
plt.legend(['nonlinear ODE','linearized'])
plt.xlabel('t')
plt.ylabel('y')
plt.title('Plot of y vs t for linearized and nonlinear ODE at y0=3' )
```

3. Let's test one more initial condition. On a single figure, plot the time response of both models when starting at an initial conditions of $y_0 = 1$. Clearly indicate which lines represent which ODEs.

```python
y0=1; #new initial condition 

y1= odeint(flowdy, y0, timesteps)
y2= odeint(lineardy, y0, timesteps)

plt.plot(timesteps,y1)
plt.plot(timesteps,y2)
plt.legend(['nonlinear ODE','linearized'])
plt.xlabel('t')
plt.ylabel('y')
plt.title('Plot of y vs t for linearized and nonlinear ODE at y0=1' )
```
